```python
# 1. Prompt the user for five numbers and store them in a list. If the largest number in the sorted list is greater than 10, print "Large"; otherwise, print "Small".

# Prompt the user for five numbers
numbers = []
for i in range(5):
    num = float(input(f"Enter number {i + 1}: "))
    numbers.append(num)

# Sort the list
numbers.sort()
print("sorted number: ",numbers)

# Check if the largest number is greater than 10
if numbers[-1] > 10:
    print("Large")
else:
    print("Small")
print("final list: ",numbers)
print("----------------------------------------")
```

```python
#  2. Ask the user for three names and append them to a list. If the list contains the name "Alice", extend it with the list ["Bob", "Charlie"]; otherwise,
#  sort the list alphabetically.

# Prompt the user for three names and store them in a list
names = []
for i in range(3):
    name = input(f"Enter name {i + 1}: ")
    names.append(name)

# Check for "Alice" in the list
if "Alice" in names:
    names.extend(["Bob", "Charlie"])
else:
    names.sort()

# Print the resulting list (optional, for verification)
print("Final list:", names)
```

```python
# 3. Have the user input two lists of numbers. If the sum of the first list is greater than the sum of the second, append the first list to the second; else, sort the combined list.

# Get two lists of numbers from the user
list1 = input("Enter numbers for the first list, separated by spaces: ")
list2 = input("Enter numbers for the second list, separated by spaces: ")

# Convert input strings to lists of floats
list1 = [float(num) for num in list1.split()]
list2 = [float(num) for num in list2.split()]

# Compare the sums and act accordingly
if sum(list1) > sum(list2):
    list2.extend(list1)  # Append first list to the second
    result = list2
else:
    result = list1 + list2
    result.sort()

# Print the result (optional)
print("Resulting list:", result)
```

**Some examples of try-except block also known as "Error Handling"**



What is try-except?

Imagine you're writing instructions for someone. Some steps might go wrong (e.g., "pick up a glass," but the glass might be slippery and fall). The try block is where you put the instructions that might cause an error or "go wrong." The except block is what you tell Python to do if that specific error happens. It's like your backup plan or how to handle the mistake gracefully instead of the whole program crashing.

```python
# 4. Get three numbers from the user and put them in a list. If the second number is even, extend the list with the first and third numbers; otherwise, append the string "Odd" to the list and then sort it.

# Get three numbers from the user
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
num3 = int(input("Enter the third number: "))

# Create a list with the three numbers
numbers = [num1, num2, num3]

# Check if the second number is even
if num2 % 2 == 0:
    # Extend the list with the first and third numbers
    numbers.extend([num1, num3])
else:
    # Append the string "Odd" and then sort the list
    numbers.append("Odd")
    # Sort only the numeric elements, then add "Odd" at the end
    nums_only = sorted([x for x in numbers if isinstance(x, int)])
    numbers = nums_only + ["Odd"]

# Output the result
print("Final list:", numbers)
```

```python
# Example 1: Converting User Input to a Number

# Scenario: You ask the user for their age, and you want to use it as a number
# (e.g., to calculate their age next year). Potential Problem: What if the user
# types "twenty" or "abc" instead of a number like "20"? The int() function will
# fail and cause a ValueError.

print("--- Example 1: Handling non-numeric input ---")
try:
    age_input = input("Please enter your age: ")
    age_number = int(age_input)  # This line might cause a ValueError
    print(f"Next year, you will be {age_number + 1} years old.")
except ValueError:
    # This code runs ONLY IF a ValueError occurs in the 'try' block
    print("Oops! That wasn't a valid number. Please enter your age using digits (e.g., 30).")
```

```python
# Example 2: Dividing by Zero

# Scenario: You want to divide one number by another. Potential Problem:
# What if the number you're dividing by (the denominator) is zero? Mathematically,
#  division by zero is undefined, and in Python, it causes a ZeroDivisionError.

print("--- Example 2: Handling division by zero ---")
numerator = 10
try:
    denominator_input = input(f"Enter a number to divide {numerator} by: ")
    denominator_number = int(denominator_input) # This could also cause a ValueError!

    result = numerator / denominator_number # This line might cause a ZeroDivisionError
    print(f"{numerator} / {denominator_number} = {result}")

except ZeroDivisionError:
    # This code runs ONLY IF a ZeroDivisionError occurs
    print("Sorry, you can't divide by zero! That's not allowed.")
except ValueError:
    # We can have multiple except blocks for different errors
    print("That wasn't a valid number for the divisor.")
```

```python
# 5. Prompt the user for a list of four words. If the list contains the word "Python", sort the list; else, append the word "Python" and then sort the list.

# Prompt the user for a list of four words
words = input("Enter four words separated by spaces: ").split()

# Check if the user entered exactly four words
if len(words) != 4:
    print("Please enter exactly four words.")
else:
    # Check for the word "Python"
    if "Python" in words:
        words.sort()
    else:
        words.append("Python")
        words.sort()

    # Print the final list
    print("Final list:", words)
```

```python
#  6. Ask the user for two numbers and create a list containing those numbers. If the first number is larger than the second, swap their positions in the list; else, append the sum of the two numbers to the list.

# Ask the user for two numbers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Create a list with the two numbers
numbers = [num1, num2]

# Check if the first number is larger than the second
if num1 > num2:
    # Swap their positions
    numbers = [num2, num1]
else:
    # Append the sum of the two numbers
    numbers.append(num1 + num2)

# Print the final list
print("Final list:", numbers)
```

```python
# 7. Have the user input a list of five colors. If "red" is in the list, sort the list alphabetically; else, append "red" to the list and then sort the list in reverse order using sorted(list, reverse=True).

# Prompt the user for a list of five colors
colors = input("Enter five colors separated by spaces: ").split()

# Check if the user entered exactly five colors
if len(colors) != 5:
    print("Please enter exactly five colors.")
else:
    # Check if "red" is in the list (case-sensitive)
    if "red" in colors:
        colors.sort()
    else:
        colors.append("red")
        colors = sorted(colors, reverse=True)

    # Print the final list
    print("Final list:", colors)
```

```python
# 8. Get a list of three fruits from the user. If the first fruit alphabetically is "apple", extend the list with ["banana", "orange"]; otherwise, append "grape" and then sort the list.

# Get a list of three fruits from the user
fruits = input("Enter three fruits separated by spaces: ").split()

# Check if exactly three fruits were entered
if len(fruits) != 3:
    print("Please enter exactly three fruits.")
else:
    # Find the first fruit alphabetically
    first_alpha = min(fruits)

    if first_alpha == "apple":
        # Extend the list with "banana" and "orange"
        fruits.extend(["banana", "orange"])
    else:
        # Append "grape" and sort the list
        fruits.append("grape")
        fruits.sort()

    # Print the final list
    print("Final list:", fruits)
```

```python
# 9. Prompt the user for two lists of two numbers each. If the sum of the elements of the first list is greater than the sum of the elements of the second list, append the second list to the first; otherwise, sort the first list.

# Prompt the user for two lists of two numbers each
list1 = list(map(int, input("Enter two numbers for the first list, separated by spaces: ").split()))
list2 = list(map(int, input("Enter two numbers for the second list, separated by spaces: ").split()))

# Check that both lists contain exactly two numbers
if len(list1) != 2 or len(list2) != 2:
    print("Each list must contain exactly two numbers.")
else:
    # Compare the sums of the two lists
    if sum(list1) > sum(list2):
        # Append the second list to the first
        list1.extend(list2)
    else:
        # Sort the first list
        list1.sort()

    # Print the final first list
    print("Final first list:", list1)
```

```python
# 10. Ask the user for their name and two favorite numbers. Create a list with the name as a string and the numbers. If the first number is greater than 5, append the string “High”; otherwise, extend the list with [0, 0, 0] and then sort it.

# Ask the user for their name
name = input("Enter your name: ")

# Ask for two favorite numbers
nums = list(map(int, input("Enter two favorite numbers separated by space: ").split()))

# Check that exactly two numbers were entered
if len(nums) != 2:
    print("Please enter exactly two numbers.")
else:
    # Create a list with the name and the two numbers
    data = [name, nums[0], nums[1]]

    # Check if the first number is greater than 5
    if nums[0] > 5:
        data.append("High")
    else:
        data.extend([0, 0, 0])
        # Sort only the numeric elements and keep the name at the start
        name_part = [data[0]]
        numbers_only = sorted([item for item in data[1:] if isinstance(item, int)])
        data = name_part + numbers_only

    # Print the final list
    print("Final list:", data)
```

```python
# 3. Have the user input two lists of numbers. Using list slicing and
# conditional expressions, create a new list containing the common
# elements of both lists if the product of the elements in the first list is
# greater than the product of the elements in the second list. Otherwise,
# create a new list by interleaving the elements of the two lists. Print the
# new list.

import math

# Get two lists of numbers from the user
list1 = list(map(int, input("Enter numbers for list 1, separated by spaces: ").split()))
list2 = list(map(int, input("Enter numbers for list 2, separated by spaces: ").split()))

# Calculate the products of the elements in both lists
product1 = math.prod(list1)
product2 = math.prod(list2)

# Create new list based on condition
if product1 > product2:
    new_list = [num for num in list1 if num in list2]
else:
    # Interleave the elements
    new_list = [val for pair in zip(list1, list2) for val in pair]
    # If lists are unequal in length, add the remaining elements
    new_list += list1[len(list2):] if len(list1) > len(list2) else list2[len(list1):]

# Print the new list
print("New list:", new_list)
```

```python
# 4. Get three numbers. Put them in a list. Using list slicing, conditional
# expressions and the sum() function, create a new list containing the
# cumulative sum of the original list if the average of the numbers is
# greater than 10. Otherwise, create a new list by replicating each element
# of the original list three times. Print the new list.

# Get three numbers from the user
numbers = [int(input(f"Enter number {i+1}: ")) for i in range(3)]

# Calculate the average
average = sum(numbers) / len(numbers)

# Conditional logic to create the new list
if average > 10:
    # Create a cumulative sum list
    new_list = [sum(numbers[:i+1]) for i in range(len(numbers))]
else:
    # Replicate each element three times
    new_list = [num for num in numbers for _ in range(3)]

# Print the new list
print("New list:", new_list)
```

```python
# 5. Prompt for a list of four words. Using list comprehensions, conditional
# expressions and string methods, create a new list containing the reversed
# versions of the words if any word contains "e". Otherwise, create a new
# list containing the lengths of the words, sorted in descending order. Print
# the new list.


# Prompt for a list of four words
words = [input(f"Enter word {i+1}: ") for i in range(4)]

# Check if any word contains the letter "e"
if any('e' in word for word in words):
    # Create a list of reversed words
    new_list = [word[::-1] for word in words]
else:
    # Create a list of word lengths, sorted in descending order
    new_list = sorted([len(word) for word in words], reverse=True)

# Print the new list
print("New list:", new_list)
```

```python
# 6. Ask for two numbers. Create a list containing those numbers. Using list
# slicing, conditional expressions, and mathematical operators, create a
# new list containing the powers of the first number (from 0 to the second
# number) if the first number is prime. Otherwise, create a new list
# containing the factors of the second number. (Assume a simple primality
# test is available without needing loops). Print the new list

import math

# Function to check if a number is prime (without using loops)
def is_prime(n):
    return n > 1 and all(n % i != 0 for i in range(2, int(math.sqrt(n)) + 1))

# Ask for two numbers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Create a list with those numbers
numbers = [num1, num2]

# Conditional logic
if is_prime(num1):
    # List of powers of num1 from 0 to num2
    new_list = [num1 ** i for i in range(num2 + 1)]
else:
    # List of factors of num2
    new_list = [i for i in range(1, num2 + 1) if num2 % i == 0]

# Print the new list
print("New list:", new_list)
```

```python
# 9. Prompt the user for two lists of two numbers each. Using list slicing,
# conditional expressions, and mathematical operators, create a new list
# containing the element-wise product of the two lists if the dot product of
# the two lists is greater than 10. Otherwise, create a new list containing
# the absolute difference between the corresponding elements of the two
# lists. Print the new list

# Prompt the user for two lists of two numbers each
list1 = list(map(int, input("Enter 2 numbers for list 1 (space-separated): ").split()))
list2 = list(map(int, input("Enter 2 numbers for list 2 (space-separated): ").split()))

# Ensure both lists have exactly two elements
if len(list1) != 2 or len(list2) != 2:
    print("Each list must contain exactly two numbers.")
else:
    # Calculate dot product
    dot_product = sum(list1[i] * list2[i] for i in range(2))

    # Conditional logic
    if dot_product > 10:
        new_list = [list1[i] * list2[i] for i in range(2)]  # Element-wise product
    else:
        new_list = [abs(list1[i] - list2[i]) for i in range(2)]  # Absolute difference

    # Print the new list
    print("New list:", new_list)
```

```python
# From a DataFrame of student marks, identify students who scored below the class average
# in at least two subjects. Columns: math, science, english.
# data = {
# 'student': ['Alice', 'Bob', 'Charlie', 'David'],
# 'math': [85, 50, 60, 90],
# 'science': [78, 65, 55, 95],
# 'english': [88, 40, 70, 85]
# }
# marks_df = pd.DataFrame(data)

import pandas as pd

# Define the DataFrame
data = {
    'student': ['Alice', 'Bob', 'Charlie', 'David'],
    'math': [85, 50, 60, 90],
    'science': [78, 65, 55, 95],
    'english': [88, 40, 70, 85]
}
marks_df = pd.DataFrame(data)

# Step 1: Compute class average for each subject
subject_avgs = marks_df[['math', 'science', 'english']].mean()

# Step 2: Compare each score to the class average
below_avg = marks_df[['math', 'science', 'english']] < subject_avgs

# Step 3: Count subjects where score is below average
below_avg['count_below_avg'] = below_avg.sum(axis=1)

# Step 4: Filter students with at least two subjects below average
result = marks_df[below_avg['count_below_avg'] >= 2]

# Show only student names
print(result[['student']])
```

```python
# Write a function that accepts a Pandas Series of strings and returns a new Series with only
# the first vowel removed from each word (case insensitive).

import pandas as pd
import re

def remove_first_vowel(series):
    def remove_vowel(word):
        # Use regex to remove only the first vowel (case insensitive)
        return re.sub(r'[aeiouAEIOU]', '', word, count=1)

    return series.apply(remove_vowel)

words = pd.Series(['Apple', 'Banana', 'Orange', 'Grape', 'Strawberry'])
new_words = remove_first_vowel(words)
print(new_words)
```

```python
# Given a NumPy array of integers, create a boolean mask that identifies prime numbers in the array

import numpy as np

# Sample input array
arr = np.array([0, 1, 2, 3, 4, 5, 10, 13, 17, 20])

# Function to check if a number is prime
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5)+1, 2):
        if n % i == 0:
            return False
    return True

# Vectorize the function
is_prime_vec = np.vectorize(is_prime)

# Create the boolean mask
prime_mask = is_prime_vec(arr)

# Output the results
print("Input array:", arr)
print("Prime mask:", prime_mask)
print("Primes in array:", arr[prime_mask])
```

```python
# Use a for loop to identify and print palindromic words from a list, ignoring case and whitespace.

words = ["Racecar", "hello", "Level", "world", "rotor", "not a palindrome", "Madam", "step on no pets"]

for word in words:
    # Remove spaces and convert to lowercase
    cleaned = word.replace(" ", "").lower()
    # Check if it's a palindrome
    if cleaned == cleaned[::-1]:
        print(f"Palindromic word: '{word}'")
```

```python
# Create a NumPy array of shape (6,6) filled with numbers from 1 to 36. Write code to extract
# all even numbers from rows with even indices only.

import numpy as np

# Create the array
arr = np.arange(1, 37).reshape(6, 6)

# Select rows with even indices (0, 2, 4)
even_index_rows = arr[::2]

# Extract even numbers from those rows
even_numbers = even_index_rows[even_index_rows % 2 == 0]

print("Original array:\n", arr)
print("\nEven numbers from rows with even indices:\n", even_numbers)
```

```python
def calculate_area(length,breadth):
  result = length * breadth
  return result


result_a = calculate_area(10,10)
print(f"The area of rectangle is {result_a}")
print("-" * 40)

result_b = calculate_area(20,20)
print(f"The area of rectangle is {result_b}")
print("-" * 40)
```

```python
def calculate_average(num1,num2,num3):
  '''
  This function calculates the average of 3 numbers by taking them as inputs
  '''
  avg_result = (num1 + num2 + num3) / 3
  return avg_result

print("Welcome to the Average Calculator..........")

input1 = float(input("Enter the first number: "))
input2 = float(input("Enter the second number: "))
input3 = float(input("Enter the third number: "))


# call the function
avg_result = calculate_average(input1,input2,input3)
avg_result_rounded = round(avg_result,2)
print(f"The calculated average is: {avg_result:.2f}")
#print(f"The average of {input1}, {input2} and {input3} is {avg_result}")
```

** Create a function that reverses the string**

```python
def reverse_string(input_str):
  '''this functions reverses the string given by the user'''
  return input_str[::-1]

reverse_string("Jingalala")
```

```python
# Develop a function get_http_status_action that takes an HTTP status code (integer) as input.

# It should return a string describing the general action or type of response:

# 200-299: 'Success'

# 300-399: 'Redirection'

# 400-499: 'Client Error'

# 500-599: 'Server Error'

# Any other code: 'Unknown Status Code Type'"



def http_status_action(status_code):
  '''
  This function determines action based on HTTP status code
  '''
  if 200 <= status_code <= 299:
    return "Success"
  elif 300 <= status_code <= 399:
    return "Redirection"
  elif 400 <= status_code <= 499:
    return "Client Error"
  elif 500 <= status_code <= 599:
    return "Server Error"
  else:
    return "Unknown Status Code Type"

# call the function

print(f"Status 200: {http_status_action(200)} ")
print(f"Status 301: {http_status_action(301)} ")
print(f"Status 404: {http_status_action(404)} ")
print(f"Status 503: {http_status_action(503)} ")
```
