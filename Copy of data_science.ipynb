**What is the total (sum) of global sales for each gaming platform ?**

```python
df.groupby('Platform')['Global_Sales'].sum()
```

**What is the total (sum) of global sales for each gaming platform ? List the top 10**

```python
global_sales_by_platform = (
    df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False)
)

print(global_sales_by_platform.head(10))
```

**What is the Global Average Sales per game for each genre ?**

```python
avg_sales_by_genre = (df.groupby('Genre')['Global_Sales'].mean().sort_values(ascending=False))

print(avg_sales_by_genre.head(10))
```
